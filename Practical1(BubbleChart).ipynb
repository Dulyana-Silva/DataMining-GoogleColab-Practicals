```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Read your CSV file
Housing_Data = pd.read_csv('/content/drive/MyDrive/Housing.csv')
# Step 3: Define bubble sizes based on 'basement'
size_map = {'no': 50, 'yes': 200}
Housing_Data['bubble_size'] = Housing_Data['basement'].map(size_map)

# Step 4: Define colors based on 'basement'
color_map = {'no': 'red', 'yes': 'blue'}
Housing_Data['bubble_color'] = Housing_Data['basement'].map(color_map)

# Step 5: Plot the bubble chart
plt.scatter(
    Housing_Data['area'],
    Housing_Data['price'],
    s=Housing_Data['bubble_size'],
    c=Housing_Data['bubble_color'],
    alpha=0.6,
    edgecolors='black'
)

# Optional formatting
plt.gca().spines[['top', 'right']].set_visible(False)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Bubble Chart: Price vs Area (Color by Parking)")
plt.show()
```
